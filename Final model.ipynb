```python
import cv2
import dlib
import face_recognition
import numpy as np

# Load pre-trained models
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks (2).dat")
face_recognizer = dlib.face_recognition_model_v1("dlib_face_recognition_resnet_model_v1.dat")

# Load a sample image to search for
search_image = face_recognition.load_image_file(r"C:\Users\saini\OneDrive\Desktop\WhatsApp Image 2024-01-31 at 13.26.14_37e76d09.jpg")
search_face_descriptor = face_recognition.face_encodings(search_image)[0]

# Open video capture from the default camera (camera index 0)
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Detect faces in the current frame
    faces = detector(frame)

    for face in faces:
        shape = predictor(frame, face)
        face_descriptor_dlib = face_recognizer.compute_face_descriptor(frame, shape)
        
        # Convert face descriptor to NumPy array
        face_descriptor_numpy = np.array(face_descriptor_dlib)

        # Compare the face descriptor with the search image
        match = face_recognition.compare_faces([search_face_descriptor], face_descriptor_numpy, tolerance=0.6)[0]

        if match:
            # Draw rectangle around the matching face
            (x, y, w, h) = (face.left(), face.top(), face.width(), face.height())
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the frame
    cv2.imshow("Video Search", frame)

    # Break the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from facenet_pytorch import MTCNN, InceptionResnetV1, extract_face


# Load pre-trained FaceNet model
device = 'cuda' if torch.cuda.is_available() else 'cpu'
mtcnn = MTCNN(keep_all=True, device=device)
model = InceptionResnetV1(pretrained='vggface2').eval().to(device)

# Load a sample image to search for
search_image = cv2.imread(r"C:\Users\saini\Downloads\WhatsApp Image 2024-02-05 at 9.17.11 PM.jpeg")
search_image_rgb = cv2.cvtColor(search_image, cv2.COLOR_BGR2RGB)

# Use MTCNN to detect faces in the search image
boxes, probs = mtcnn.detect(search_image_rgb)

if boxes is not None:
    # Extract faces and calculate embeddings
    faces = extract_face(search_image_rgb, boxes)
    embeddings = model(faces)

# Open video capture from the default camera (camera index 0)
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Use MTCNN to detect faces in the current frame
    boxes, probs = mtcnn.detect(frame)

    if boxes is not None:
        # Extract faces and calculate embeddings
        faces = extract_face(frame, boxes)
        embeddings_frame = model(faces)

        # Compare embeddings with the search image
        for i, embedding_frame in enumerate(embeddings_frame):
            match = np.linalg.norm(embedding_frame.detach().numpy() - embeddings.detach().numpy()) < 0.6

            if match:
                # Draw rectangle around the matching face
                (x, y, w, h) = (int(boxes[i][0]), int(boxes[i][1]), int(boxes[i][2] - boxes[i][0]), int(boxes[i][3] - boxes[i][1]))
                cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Display the frame
    cv2.imshow("Video Search", frame)

    # Break the loop when 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object
cap.release()
cv2.destroyAllWindows()
```
